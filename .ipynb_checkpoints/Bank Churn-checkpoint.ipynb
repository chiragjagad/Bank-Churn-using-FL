```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("BankChurners.csv")
dataset.T
```

```python
dataset.info()
```

```python
dataset.drop('Unnamed: 21', axis=1, inplace=True)
dataset.T
```

```python
dataset
```

```python
dummies_dataset = pd.get_dummies(dataset[["Attrition_Flag", "Gender"]])
dummies_dataset
```

```python
dummies_dataset.drop(['Attrition_Flag_Existing Customer', "Gender_F"], axis=1, inplace=True)
dummies_dataset = dummies_dataset.rename(columns={"Attrition_Flag_Attrited Customer": "Attrited Customer", "Gender_M" : "Gender"})
dummies_dataset
```

```python
dataset.drop(['Attrition_Flag', "Gender", "CLIENTNUM"], axis=1, inplace=True)
```

```python
dataset = pd.concat((dummies_dataset, dataset), axis=1)
dataset
```

```python
dataset.hist(figsize=[20,20])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(dataset.corr(method='pearson'), cmap="RdBu", annot=True,ax=ax, vmin=-1, vmax=1, center=0, fmt=".2f")
```

```python
for col in dataset.columns.difference(dataset._get_numeric_data().columns):
    print(dataset[col].value_counts())
    print('#'*50)
```

```python
def bar_chart(column):
    attrited=dataset[dataset["Attrited Customer"]==1][column].value_counts()
    exisitng=dataset[dataset["Attrited Customer"]==0][column].value_counts()
    df=pd.DataFrame([attrited,exisitng])
    df.index=["Attrited","Exsiting"]
    df.plot(kind="bar",figsize=(8,4))
bar_chart('Card_Category')
bar_chart('Education_Level')
bar_chart('Income_Category')
bar_chart('Marital_Status')
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
dataset['Card_Category'] = oe.fit_transform(dataset[['Card_Category']])
dataset
```

```python
dummies_dataset1 = pd.get_dummies(dataset[["Marital_Status", "Income_Category", "Education_Level"]])
dummies_dataset1
```

```python
dataset.drop(['Marital_Status', "Income_Category", "Education_Level"], axis=1, inplace=True)
dataset = pd.concat((dataset, dummies_dataset1), axis=1)
dataset
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy='minority', random_state=0)
oversampled_datasetX, oversampled_datasetY = sm.fit_resample(dataset.drop('Attrited Customer', axis=1), dataset['Attrited Customer'])
oversampled_dataset = pd.concat([pd.DataFrame(oversampled_datasetY), pd.DataFrame(oversampled_datasetX)], axis=1)
oversampled_dataset
```

```python
oversampled_dataset['Attrited Customer'].value_counts()
```

```python
y = oversampled_dataset['Attrited Customer']
oversampled_dataset.drop(['Attrited Customer'], axis = 1, inplace = True)
oversampled_dataset
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_dataset = pd.DataFrame(sc.fit_transform(oversampled_dataset),columns = oversampled_dataset.columns)
scaled_dataset
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(scaled_dataset, y, test_size=0.2, random_state=0)
```

```python
import pickle
with open('xtrain.pkl','wb') as f:
    pickle.dump(X_train, f)
```

```python
with open('xtest.pkl','wb') as f:
    pickle.dump(X_test, f)
with open('ytrain.pkl','wb') as f:
    pickle.dump(y_train, f)
with open('ytest.pkl','wb') as f:
    pickle.dump(y_test, f)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
pred=rfc.predict(X_test)
pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
from sklearn.model_selection import cross_val_score
cv = cross_val_score(rfc,X_test,y_test,cv = 10)
print(cv)
print(cv.mean())
```

```python
import pickle
with open("xtrain.pkl", "rb") as f:
    X_train = pickle.load(f)
with open("xtest.pkl", "rb") as f:
    X_test = pickle.load(f)
with open("ytrain.pkl", "rb") as f:
    y_train = pickle.load(f)
with open("ytest.pkl", "rb") as f:
    y_test = pickle.load(f)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential(
    [
    Dense(16, input_shape=(33,), activation=LeakyReLU(alpha=0.01)),
    Dropout(0.1),
    Dense(16, activation="relu"),
    Dropout(0.1),
    Dense(8, activation="relu"),
    Dropout(0.1),
    Dense(1, activation="sigmoid"),
    ]
)
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, batch_size=320, epochs=100)
```

```python
logits = model.predict(X_test)
```

```python
bce = tf.keras.losses.BinaryCrossentropy(from_logits=True)
loss = bce(y_test, logits.squeeze())
loss
```

```python
y_pred = (logits > 0.5)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_pred.squeeze(), y_test)
acc
```
